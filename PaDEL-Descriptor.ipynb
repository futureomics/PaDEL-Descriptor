```python
import glob
xml_files = glob.glob("*.xml") #return file paths named .xml
xml_files.sort()
xml_files
```

```python
#create a list and assign them to the xml files( to compute descriptors)
FP_list = ['AtomPairs2DCount',
 'AtomPairs2D',
 'EState',
 'CDKextended',
 'CDK',
 'CDKgraphonly',
 'KlekotaRothCount',
 'KlekotaRoth',
 'MACCS',
 'PubChem',
 'SubstructureCount',
 'Substructure']
```

```python
fp = dict(zip(FP_list, xml_files)) #create a dictionary of all..
fp
```

```python
fp['AtomPairs2D']
```

```python
from padelpy import padeldescriptor   #import downloaded padel descriptor
df2 = pd.concat( [df['canonical_smiles'],df['molecule_chembl_id']], axis=1 )  #extract smiles and cid columns from dataframe and add to df2
df2.to_csv('/content/molecule.smi', sep='\t', index=False, header=False) #save to csv file without headers and index
df2.to_csv('/content/molecule2.smi', sep='\t', index=False, header=False)
df2
```

```python
from padelpy import padeldescriptor

fingerprint = 'PubChem' #computing pubchem descriptors

fingerprint_output_file = ''.join([fingerprint,'.csv']) #Compute descriptor using [DescriptorName.csv]eg PubChem.csv or Substructure.csv
fingerprint_descriptortypes = fp[fingerprint]

padeldescriptor(mol_dir='/content/molecule2.smi',
                d_file=fingerprint_output_file, #'Substructure.csv'
                #descriptortypes='SubstructureFingerprint.xml',
                descriptortypes= fingerprint_descriptortypes,
                detectaromaticity=True,
                standardizenitro=True,
                standardizetautomers=True,
                threads=2,
                removesalt=True,
                log=True,
                fingerprints=True)
```
